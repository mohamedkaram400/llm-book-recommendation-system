### Import libraries

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv

load_dotenv()
```

### Read data

```python
import pandas as pd

books = pd.read_csv('../data/processed/books_cleand.csv')
```

```python
books.head()
```

### Save tagged_description as text for splitting it 

```python
books['tagged_description'].to_csv('../data/external/tagged_description.txt', sep="\n", index=False, header=False)
```

### Load and split text

```python
row_documents = TextLoader('../data/external/tagged_description.txt').load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(row_documents)
```

```python
documents[0]
```

### # Store in Chroma and Use a Free Embedding Model from Hugging Face

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

# Use a Free Embedding Model from Hugging Face
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Store Documents in ChromaDB
db_books = Chroma.from_documents(documents, embedding=embedding_model)
```

### Write search query for testing

```python
query = "a book to teach childern about nature"
docs = db_books.similarity_search(query, k=10)

docs
```

```python
isbn_extracted = docs[0].page_content.split()[0].strip().rstrip(":") 
books[books['isbn13'] == int(isbn_extracted)]
```

```python
def retrieve_semantic_recommendations(query: str, top_k: int = 10) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0].strip(":"))]

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
